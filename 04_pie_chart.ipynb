```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [100,700,50,20]
exp_labels = ['CNG','BUS','FOOD','RICKSHAW']
```

```python
# autopct show the percentage
# explode array values defines how much to get out or explode in the chart
# startangle is the starting angle
plt.pie(exp_vals,labels=exp_labels,radius=1, autopct='%0.2f%%',explode=[0,0,0.2,0],startangle=180)
# in terms of pixels they are different but equal makes it like a perfect circle
plt.axis('equal')
# to get rid of extra info


# saving a matplotlib
plt.savefig('piechart.png',bbox_inches='tight', pad_inches=2, transparent=True)
```
