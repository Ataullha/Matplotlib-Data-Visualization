```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# in histogram we need single dimension data y axis contains the frequency
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]

# by default 10 bins (ranges)
# rwidth means the relative width of the bin
# plt.hist(blood_sugar, bins=3, rwidth=0.95)
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.75, color='g', histtype='step')
```

```python
blood_sugar_man = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_woman = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_man,blood_sugar_woman],bins=[80,100,125,150],color=['g','r'],label=['man','woman'],orientation='vertical')
plt.legend()
plt.xlabel('SUGAR RANGE')
plt.ylabel('TOTAL PATIENTS')
plt.title('BLOOD SUGAR LEVEL')
```
