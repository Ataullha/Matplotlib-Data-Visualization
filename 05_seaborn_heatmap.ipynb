```python
import seaborn as sns

import matplotlib.pyplot as plt
import numpy as np
```

```python
cars = sns.load_dataset('mpg')
```

```python
cars.head()
```

```python
cars.groupby('origin').cylinders.value_counts()
```

```python
cars.groupby('origin').cylinders.value_counts().unstack()
```

```python
origin_cyl = cars.groupby('origin').cylinders.value_counts().unstack().fillna(0)
```

```python
sns.heatmap(origin_cyl)
```

# Color Palette

```python
sns.heatmap(
    origin_cyl,
    cmap='Blues'
           )
```

```python
sns.heatmap(
    origin_cyl,
    cmap='RdBu'
           )
```

```python
sns.heatmap(
    cars.corr(),
    cmap='Blues',
    center=0,
    vmin=-1,
    vmax=1
           )
```

```python
sns.heatmap(
#     cars correlation
    cars.corr(),
#     color map
    cmap='Blues',
#     set the center to 0
    center=0,
#     value min
    vmin=-1,
#     value max
    vmax=1,
#     annotation True
    annot=True,
#     format
    fmt="0.0f"
           )
```

```python
labels = ['MPG', 'CYLINDERS', 'DISPLACEMENT', 'HORSEPOWER', 'WEIGHT', 'ACCELERATION', 'MODEL_YEAR']
```

# All in One

```python
sns.heatmap(
    cars.corr(),
    cmap='RdBu',
    center=0,
    vmin=-1,
    vmax=1,
    annot=True,
    fmt="0.0f",
    linewidth=1,
    linecolor='green',
    xticklabels=labels,
#     can change the label like previous one or can pass True/False
    yticklabels=False,
#     if you want each of the little box to be square
    square=True,
    annot_kws={
       'fontsize':16,
       'fontweight': 'bold',
        'fontfamily': 'serif',
#         not prefarable because heatmap do it in an itteligent way
        'color':'black'
    }
           );
```
