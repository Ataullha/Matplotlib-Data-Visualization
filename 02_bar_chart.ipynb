```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['GOOGL','AMZN','NVIDIA','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
xpos = np.arange(len(company))
```

```python
xpos
```

```python
# replace ypos with company list
plt.xticks(ypos,company)
# xpos is the position of x but here we are plotting two charts so we -0.2 and +0.2 to make a balance in order to keep the width 0.4
plt.bar(xpos-0.2,revenue,width=0.4,label='REVENUE')
plt.bar(xpos+0.2,profit,width=0.4,label='PROFIT')
plt.xlabel('COMPANIES')
plt.ylabel('IN BILLION DOLLARS')
plt.title('US TECH STOCKS')

plt.legend()
```

```python
plt.yticks(ypos,company)
plt.barh(xpos-0.2,revenue,label='REVENUE')
plt.barh(xpos+0.2,profit,label='PROFIT')
plt.xlabel('COMPANIES')
plt.ylabel('IN BILLION DOLLARS')
plt.title('US TECH STOCKS')

plt.legend()
```
