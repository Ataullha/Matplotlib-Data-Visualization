```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# last seven days weather records
x = [1,2,3,4,5,6,7]
y = [29,32,33,32,33,31,29]
```

```python
plt.ylabel('Temparature')
plt.xlabel('Week')
plt.title('Weekly Temparature Update')
# g means green
plt.plot(x,y,color='g',linestyle='dashed',linewidth=5)
```

```python
# r means red -.-.-. is the line * is the marker alpha is the transperency
plt.plot(x,y,'r-.',marker='*',markersize=20,markerfacecolor='b',alpha=0.2)
```

```python
x = [1,2,3,4,5,6,7]
temp_high = [29,32,33,32,33,31,29]
temp_low = [27,30,31,30,31,29,27]
temp_avg = [28,31,32,31,32,30,28]
```

```python
plt.ylabel('Temparature')
plt.xlabel('Days')
plt.title('Weather')

plt.plot(x,temp_high,label='High')
plt.plot(x,temp_avg,label='AVG')
plt.plot(x,temp_low,label='LOW')

# legend loc means location shadow means legends shadow
plt.legend(loc='best',shadow='true',fontsize='small')

plt.grid()
```
